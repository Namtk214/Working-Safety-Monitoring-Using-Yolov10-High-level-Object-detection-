```python
from ultralytics import YOLOv10
MODEL_PATH = 'yolov10n.pt' # Just the path to the model file
model = YOLOv10(MODEL_PATH) # Load the model using the path
```

```python
YAML_PATH = '/content/safety_helmet_dataset/data.yaml'
EPOCHS = 50
BATCH_SIZE = 640
IMG_SIZE = 256

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            batch=BATCH_SIZE,
            imgsz=IMG_SIZE)
```

DEMO RESULT

```python
from google.colab.patches import cv2_imshow
TRAINED_MODEL_PATH = '/content/yolov10/yolov10/yolov10/yolov10/runs/detect/train2/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)
IMG_URL = 'https://lsvn.vn/storage/uploads/photos/1015/60581402658f9.jpg'
CONF_THRESHOLD = 0.3
results = model.predict(source=IMG_URL,imgsz = IMG_SIZE, conf=CONF_THRESHOLD)
annotated_img = results[0].plot()
cv2_imshow(annotated_img)
```
